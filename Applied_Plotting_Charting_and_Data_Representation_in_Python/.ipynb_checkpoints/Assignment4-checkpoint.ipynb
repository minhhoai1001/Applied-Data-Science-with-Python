## Data 
1. [WHO Coronavirus (COVID-19) Dashboard](https://covid19.who.int/table)
2. Our world in data: [Data on COVID-19 (coronavirus) vaccinations](https://github.com/owid/covid-19-data/blob/master/public/data/vaccinations/vaccinations.csv)
3. [2021 World Population by Country](https://worldpopulationreview.com/)

```python
%matplotlib notebook

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load data vaccin

```python
df_vaccin = pd.read_csv('data/vaccinations.csv')
ds_vaccin = df_vaccin.groupby(['location']).max()
ds_vaccin = ds_vaccin[['total_vaccinations', 'people_vaccinated']]
ds_vaccin.head(10)
```

## Load data covid

```python
df_covid = pd.read_csv('data/WHO COVID-19 global table data March 20th 2021 at 9.40.21 AM.csv')
df_covid =df_covid.drop(0).sort_values(['Cases - cumulative total'],ascending=False)
df_covid_10 = df_covid.iloc[0:10]

df_covid_10 = df_covid_10[['Name', 'Cases - cumulative total', 'Deaths - cumulative total']]

df_covid_10=df_covid_10.rename(columns={'Cases - cumulative total': 'Cases', 
                                        'Deaths - cumulative total': 'Deaths'})
    
df_covid_10['Name'] = df_covid_10['Name'].replace({'United States of America': 'United States',
                                                   'Russian Federation': 'Russia',
                                                    'The United Kingdom': 'United Kingdom'})
df_covid_10.set_index('Name', inplace=True)
df_covid_10.head(10)
```

## Load data 2021 World Population

```python
df_popula = pd.read_csv('data/World Population.csv')
df_popula_21 = df_popula[['name','pop2021']]
df_popula_21.set_index('name', inplace=True)
df_popula_21.head(10)
```

## Merge three data and prepare data

```python
data_10 = df_covid_10.merge(ds_vaccin, how='inner', left_index=True, right_index=True)
data_10 = data_10.merge(df_popula_21, how='inner', left_index=True, right_index=True)
data_10.reset_index(inplace=True)

data_10['Percent_vaccin'] = data_10['people_vaccinated']/(data_10['pop2021']*10)
data_10['Cases'] = data_10['Cases']/100000
data_10['Deaths'] = data_10['Deaths']/100000
data_10.head(10)
```

```python
tidy = data_10.melt(id_vars='index', value_vars=['Cases', 'Deaths'])
tidy.rename(columns={'index': 'Country', 'value':'Units: 100,000 people', 'variable':'Cases or Deaths'}, inplace=True)

percent = data_10.melt(id_vars='index', value_vars=['Percent_vaccin'])
percent.rename(columns={'index': 'Country', 'value':'Percentage'}, inplace=True)
tidy.head(10)
```

## Plot data

```python
sns.set_style('darkgrid')
fig, ax1 = plt.subplots(figsize=(10, 10))

splot = sns.barplot(x='Country', y='Units: 100,000 people', hue='Cases or Deaths', data=tidy, ax=ax1)

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    
ax2 = ax1.twinx()

line = sns.lineplot(data = percent, marker='o', palette=['red'], sort = False, ax=ax2)

ax2.set(ylabel='Percent people has vaccinated per popular (%)')
ax2.legend(bbox_to_anchor=(0.5, 1), loc=2)
plt.title('The Countries With The Most COVID-19 Cases and Vaccine')
```
